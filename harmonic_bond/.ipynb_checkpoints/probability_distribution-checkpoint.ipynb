This notebook will show how to determine the probability distribution for a harmonic bond.

```python
# import some libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

The generic form for a harmonic bond: <br>
$U(x) = \frac{1}{2}k\left(x-x_0\right)^2$

```python
x0 = 0.0
k = 1.0
x = np.arange(-10,10,0.1)
plt.plot(x,0.5*k*(x-x0)**2)
```

If this is the energy function for our harmonic bond, the equilibrium value is 0 and we expect to observe very few molecules with very small or large values of the bond distance.

We can compute the probability of observing a particular x as: <br>

$P(x) = \frac{\exp(-U(x))}{\int_{-\infty}^{\infty} \exp(-U(x)) dx}$ <br>

Or: <br>

$P(x) \propto \exp(-U(x))$


```python
plt.plot(x,np.exp(-0.5*k*(x-x0)**2))
```

What does this look like?

A guassian distribution has a generic form: <br>

$P(x) \propto \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$ <br>

where $\mu$ is the mean and $\sigma$ is the standard deviation of the distribution.
